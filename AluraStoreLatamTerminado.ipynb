### Importación de datos



```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from matplotlib.patches import Patch



url = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_1%20.csv"
url2 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_2.csv"
url3 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_3.csv"
url4 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_4.csv"

tienda = pd.read_csv(url)
tienda2 = pd.read_csv(url2)
tienda3 = pd.read_csv(url3)
tienda4 = pd.read_csv(url4)

#Establecemos la lista "tiendas[]", para recorrer cada una de las listas de la tiendas
tiendas = [tienda, tienda2, tienda3, tienda4]
nombres_tiendas = ["Tienda1", "Tienda2", "Tienda3", "Tienda4"]

dfs_unificados = []
for df, nombre in zip(tiendas, nombres_tiendas):
    df["Tienda"] = nombre
    dfs_unificados.append(df)

df = pd.concat(dfs_unificados, ignore_index=True)

df.head()
```

#1. Análisis de facturación



```python
# sumar la columna precio de cada una de las tiendas y mostrar el total de cada tienda independiente

#Utilizamos la funcion for para recorrer con total_tiendas en "tiendas[]",
#Utilizamos la funcion sum para sumar los valores en la lista de "Precio", finalmente imprimimos

totales = []
nombres_tiendas = []
for i, total_tiendas in enumerate(tiendas):
    total = sum(total_tiendas["Precio"])
    print(f"Tienda {i+1}: {total:,.2f} COP")  # Mantenemos tu salida original
    totales.append(total)
    nombres_tiendas.append(f"Tienda {i+1}")
```

```python
plt.figure(figsize=(10, 6))

# Configuración del gráfico
bars = plt.bar(nombres_tiendas, totales,
               color=['#3498db', '#2ecc71', '#e74c3c', '#f39c12'],
               edgecolor='black', linewidth=1.2)

# Añadir los valores exactos encima de cada barra
for bar in bars:
    height = bar.get_height()
    plt.text(bar.get_x() + bar.get_width()/2., height,
             f'${height:,.2f} COP',
             ha='center', va='bottom', fontsize=10, fontweight='bold')

# Personalización del gráfico
plt.title('Ingresos Totales por Tienda', pad=20, fontsize=14, fontweight='bold')
plt.xlabel('Tienda', fontsize=12)
plt.ylabel('Total Ventas (COP)', fontsize=12)
plt.xticks(fontsize=10)
plt.yticks(fontsize=10)

# Formato del eje Y en miles/millones
plt.gca().yaxis.set_major_formatter(plt.matplotlib.ticker.StrMethodFormatter('${x:,.0f}'))

# Grid y bordes
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.gca().spines['top'].set_visible(False)
plt.gca().spines['right'].set_visible(False)

plt.tight_layout()
plt.show()

# Encontrar índice de la tienda con mayores ventas
indice_max = totales.index(max(totales))

# Colores rojizos para las demás tiendas
colores_rojizos = ['#e57373', '#ef5350', '#f44336', '#e53935', '#d32f2f', '#c62828', '#b71c1c']
colores = []

# Asignar colores
for i in range(len(totales)):
    if i == indice_max:
        colores.append('#2ecc71')  # Verde para la mayor
    else:
        colores.append(colores_rojizos[i % len(colores_rojizos)])  # Rojizos diferentes

# Explode para separar la tienda con mayores ventas
explode = [0] * len(totales)
explode[indice_max] = 0.1  # Separar solo la mayor

# Crear gráfico de pastel
plt.figure(figsize=(8, 8))
plt.pie(totales, labels=nombres_tiendas, autopct='%1.1f%%',
        colors=colores, explode=explode, startangle=140, shadow=True)

# Título
plt.title('Distribución de Ventas por Tienda (Gráfico de Pastel)', fontsize=14, fontweight='bold', pad=20)

plt.tight_layout()
plt.show()



```

# 2. Ventas por categoría

```python
#calcular la cantidad de productos vendidos por categoría en cada tienda.
#La idea es agrupar los datos por categoría y contar el número de ventas de cada tipo, mostrando las categorías más populares de cada tienda.

# 1. Calcular productos vendidos por categoría en cada tienda
ventas_por_categoria = df.groupby(['Tienda', 'Categoría del Producto']).size().reset_index(name='Total_Ventas')

# 2. Ordenar por tienda y ventas (de mayor a menor)
ventas_ordenadas = ventas_por_categoria.sort_values(by=['Tienda', 'Total_Ventas'], ascending=[True, False])

# 3. Mostrar las categorías más populares por tienda (top 1 de cada tienda)
top_categorias = ventas_ordenadas.groupby('Tienda').head(1)
```

```python
# Usamos 'ventas_por_categoria', que ya agrupa las ventas por tienda y categoría
plt.figure(figsize=(12, 6))

# Crear gráfico
plt.figure(figsize=(12, 6))
sns.barplot(data=ventas_por_categoria, x='Tienda', y='Total_Ventas', hue='Categoría del Producto', palette='Set2')

# Títulos y etiquetas
plt.title('Ventas por Categoría en Cada Tienda', fontsize=14, fontweight='bold', pad=20)
plt.xlabel('Tienda', fontsize=12)
plt.ylabel('Cantidad Vendida', fontsize=12)
plt.xticks(rotation=45)

# Leyenda fuera del gráfico
plt.legend(title='Categoría', bbox_to_anchor=(1.05, 1), loc='upper left')

# Estética
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.gca().spines['top'].set_visible(False)
plt.gca().spines['right'].set_visible(False)

plt.tight_layout()
plt.show()
```

# 3. Calificación promedio de la tienda


```python
#Calificaciones promedio de los cliente para cada tienda, es decir, conocer la satisfacción del cliente con los productos vendidos

# 4. Calcular calificaciones promedio por tienda (satisfacción del cliente)
calificaciones_promedio = df.groupby('Tienda')['Calificación'].mean().reset_index()
calificaciones_promedio.columns = ['Tienda', 'Calificación Promedio']

# Opcional: Redondear a 2 decimales para mejor presentación
calificaciones_promedio['Calificación Promedio'] = calificaciones_promedio['Calificación Promedio'].round(2)
```

```python
# --- GRÁFICO DE BARRAS HORIZONTAL  Y  PASTEL ---

plt.figure(figsize=(8, 5))

# Colores personalizados (opcional)
colores = ['#3498db', '#2ecc71', '#e74c3c', '#f39c12']

# Crear gráfico horizontal
bars = plt.barh(calificaciones_promedio['Tienda'],
                calificaciones_promedio['Calificación Promedio'],
                color=colores, edgecolor='black', linewidth=1.2)

# Añadir etiquetas al final de cada barra
for bar in bars:
    width = bar.get_width()
    plt.text(width + 0.05, bar.get_y() + bar.get_height() / 2,
             f'{width:.2f}', va='center', ha='left', fontsize=10, fontweight='bold')

# Títulos y etiquetas
plt.title('Calificación Promedio por Tienda', fontsize=14, fontweight='bold', pad=15)
plt.xlabel('Calificación Promedio', fontsize=12)
plt.ylabel('Tienda', fontsize=12)
plt.xlim(0, 5)  # Escala de 0 a 5 para claridad

# Estética
plt.grid(axis='x', linestyle='--', alpha=0.6)
plt.gca().spines['top'].set_visible(False)
plt.gca().spines['right'].set_visible(False)

# Datos
labels = calificaciones_promedio['Tienda']
sizes = calificaciones_promedio['Calificación Promedio']
colors = ['#3498db', '#2ecc71', '#e74c3c', '#f39c12']  # mismos colores que en el gráfico de barras
explode = [0.05 if val == max(sizes) else 0 for val in sizes]  # resaltar la tienda con mayor calificación

# Crear figura Pastel
fig, ax = plt.subplots(figsize=(8, 6))

# Gráfico de pastel
wedges, texts, autotexts = ax.pie(
    sizes, labels=None, colors=colors, autopct='%1.2f%%', startangle=140,
    explode=explode, shadow=True, wedgeprops={'edgecolor': 'black'})

# Añadir título
ax.set_title('Distribución de Calificaciones Promedio por Tienda', fontsize=14, fontweight='bold', pad=20)

# Añadir leyenda fuera del gráfico (tabla de colores)
ax.legend(wedges, labels, title="Tiendas", loc="center left", bbox_to_anchor=(1, 0.5), fontsize=10, title_fontsize=11)

plt.tight_layout()
plt.show()
```

# 4. Productos más y menos vendidos

```python
# 1. Calcular frecuencia de ventas por producto en cada tienda
ventas_por_producto = df.groupby(['Tienda', 'Producto']).size().reset_index(name='Ventas')

# 2. Encontrar top 5 y bottom 5 productos por tienda
top_productos = ventas_por_producto.groupby('Tienda').apply(lambda x: x.nlargest(5, 'Ventas')).reset_index(drop=True)
top_productos['Tipo'] = 'Más vendidos'

bottom_productos = ventas_por_producto.groupby('Tienda').apply(lambda x: x.nsmallest(5, 'Ventas')).reset_index(drop=True)
bottom_productos['Tipo'] = 'Menos vendidos'

# 3. Unir los resultados
productos_combinados = pd.concat([top_productos, bottom_productos], ignore_index=True)
```

```python
# 4. Visualización
sns.set(style="whitegrid")
tiendas_unicas = productos_combinados['Tienda'].unique()

for tienda in tiendas_unicas:
    datos_tienda = productos_combinados[productos_combinados['Tienda'] == tienda]

    plt.figure(figsize=(10, 6))
    colores = datos_tienda['Tipo'].map({'Más vendidos': '#2ecc71', 'Menos vendidos': '#e74c3c'})
    bars = plt.barh(datos_tienda['Producto'], datos_tienda['Ventas'], color=colores, edgecolor='black')

    plt.title(f"Productos Más y Menos Vendidos - {tienda}", fontsize=14, fontweight='bold')
    plt.xlabel("Cantidad Vendida")
    plt.ylabel("Producto")
    plt.grid(axis='x', linestyle='--', alpha=0.7)

    # Mostrar valores al final de cada barra
    for bar in bars:
        ancho = bar.get_width()
        plt.text(ancho + 1, bar.get_y() + bar.get_height() / 2, f'{int(ancho)}', va='center')

    # Agregar leyenda personalizada
    legend_elements = [
        Patch(facecolor='#2ecc71', edgecolor='black', label='Más vendidos'),
        Patch(facecolor='#e74c3c', edgecolor='black', label='Menos vendidos')
    ]
    plt.legend(handles=legend_elements, title='Tipo de Producto', loc='upper right')

    plt.tight_layout()
    plt.show()
```

# 5. Envío promedio por tienda

```python
# 1. Costo de envío promedio por tienda
costo_promedio = df.groupby('Tienda')['Costo de envío'].mean().round(2).reset_index()
costo_promedio.columns = ['Tienda', 'Costo de envío promedio (COP)']

# 2. Porcentaje del costo de envío respecto al precio
df['% costo envío'] = (df['Costo de envío'] / df['Precio'] * 100).round(2)
porcentaje_promedio = df.groupby('Tienda')['% costo envío'].mean().round(2).reset_index()

# Combinar ambos resultados
resultados_completos = pd.merge(costo_promedio, porcentaje_promedio, on='Tienda')
```

```python
# Asignar colores vivos
colores_barras = sns.color_palette("bright", len(resultados_completos))
colores_pastel = sns.color_palette("bright", len(resultados_completos))

# ========== GRÁFICO DE BARRAS ==========
plt.figure(figsize=(10, 6))
bar_width = 0.4
index = range(len(resultados_completos))

# Barras de costo promedio
for i, valor in enumerate(resultados_completos['Costo de envío promedio (COP)']):
    plt.bar(i, valor, width=bar_width, color=colores_barras[i])
    plt.text(i, valor + 500, f"${valor:,.0f}", ha='center', va='bottom', fontsize=10, weight='bold')

plt.xlabel('Tienda')
plt.ylabel('Costo de envío promedio (COP)')
plt.title('Costo de Envío Promedio por Tienda')
plt.xticks(index, resultados_completos['Tienda'])
plt.tight_layout()
plt.show()


# ========== GRÁFICO DE PASTEL ==========
menor_valor = resultados_completos['Costo de envío promedio (COP)'].min()
explode = [0.1 if valor == menor_valor else 0 for valor in resultados_completos['Costo de envío promedio (COP)']]

plt.figure(figsize=(8, 6))
plt.pie(
    resultados_completos['Costo de envío promedio (COP)'],
    labels=resultados_completos['Tienda'],
    autopct=lambda p: f"${p * sum(resultados_completos['Costo de envío promedio (COP)']) / 100:,.0f}",
    colors=colores_pastel,
    explode=explode,
    startangle=140,
    wedgeprops=dict(edgecolor='white')
)
plt.title('Distribución del Costo de Envío Promedio por Tienda')
plt.tight_layout()
plt.show()
```
